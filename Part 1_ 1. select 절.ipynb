# SQL 이란
- Stucture Query Language (구조적 질의 언어)
- 데이터를 검색하고 조작하는 언어

# SQL의 종류
1. Query 문: 데이터를 검색하는 언어 => select 문의 6가지 절


2. DML 문: Data Manipulation Language (조작)  
    insert: 데이터 입력 언어  
    update: 데이터 수정 언어  
    Delete: 데이터 삭제 언어  
    Merge: 데이터 입력, 수정, 삭제를 한번에 수행하는 명령어  
    
    
3. DDL 문: Data Definition Language  
    Create: 생성  
    Alter: 수정  
    Drop: 삭제  
    Truncate: 삭제  
    Rename: 이름 변경  
      
      
4. DCL 문: Data Control Language  
    Grant: 데이터를 접근할 수 있는 권한 부여  
    Revoke: 데이터를 접근할 수 있는 권한을 취소  
    
    
5. TCL 문: Transaction Control Language  
    Commit: 현재의 데이터의 상태를 DB에 영구히 저장  
    Rollback: 지금까지 했던 작업들을 다 취소하겠다.  
    Savepoint: 특정 시점까지 롤백하는 기능

# 컬럼 별칭을 사용하여 출력되는 컬럼명 변경
"컬럼 별칭은 컬럼명 대신에 다른 컬럼명을 지정할 때 사용하는 문법"  

※ 컬럼명 뒤에 as를 쓰고 컬럼별칭을 작성한다.  
- 결과가 출력될 때 컬럼 별칭이 출력됨  
- as는 생략가능  
- 원본 데이터에서는 수정이 되지 않고 출력되는 데이터에서만 수정

<br>

※ 컬럼 별칭에 대소문자를 구분하고 싶거나 공백문자나 특수문자를 넣으려면 양쪽에 더블 퀘테이션마크(" ")를 둘러줘야 함

문제 8. 이름과 입사일, 부서번호를 출력하는데 컬럼명이 한글로 이름, 입사일, 부서번호로 출력되게 하시오

```python
%%sql

select ename as 이름, hiredate as 입사일, deptno as 부서번호
    from emp
```

문제 9. 이름과 월급을 출력하는데 컬럼명이 아래와 같이 출력되게 하라

# 연결 연산자 ( || )

"연결 연산자는 두 컬럼의 데이터를 연결해서 출력하는 연산자"

문제 10. 아래와 같이 결과를 출력하시오  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"KING의 직업은 PRESIDENT 입니다"  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"SCOTT의 직업은 ANALYST 입니다"

```python
%%sql

select ename||'의 직업은 '|| job||' 입니다'
    from emp
```

문제 11. 위의 결과에서 출력되 컬럼명을 사원정보라는 한글로 출력 되게 하시오!

```python
%%sql

select ename||'의 직업은 '||job||' 입니다' as "사원정보"
    from emp
```
